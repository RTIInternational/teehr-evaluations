```python
import teehr

from bokeh.io import output_notebook
output_notebook()
```

```python
# Create an Evaluation object to read from S3 bucket.
ev = teehr.Evaluation(dir_path="s3a://ciroh-rti-public-data/teehr-data-warehouse/v0_4_evaluations/e1_camels_daily_streamflow")
```

```python
ev.primary_timeseries.filter("location_id = 'usgs-01047000'").to_pandas().teehr.timeseries_plot()
```

```python
sec_ids = ev.location_crosswalks.filter("primary_location_id = 'usgs-01047000'").to_pandas().secondary_location_id
ev.secondary_timeseries.filter(
    {
        "column": "location_id",
        "operator": "in",
        "value": sec_ids
    }
).to_pandas().teehr.timeseries_plot()
```

```python
ev.spark.stop()
```
