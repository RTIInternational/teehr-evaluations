# Standard Camels Hourly Streamflow Evaluation

```python
# import teehr
import teehr
from bokeh.io import output_notebook

# Tell Bokeh to output plots in the notebook
output_notebook()
```

```python
# Create an Evaluation object to read from S3 bucket.
ev = teehr.Evaluation(dir_path="s3a://ciroh-rti-public-data/teehr-data-warehouse/v0_4_evaluations/e3_usgs_hourly_streamflow")
```

```python
%%time
ev.metrics.query(
    order_by=["primary_location_id"],
    group_by=["primary_location_id"],
    include_metrics=[
        teehr.Metrics.KlingGuptaEfficiency(),
        teehr.Metrics.NashSutcliffeEfficiency(),
        teehr.Metrics.RelativeBias()
    ]
).to_pandas()
```

```python
ev.spark.stop()
```
